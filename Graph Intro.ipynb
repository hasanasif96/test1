```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\hasan\Downloads\Updated data files\Updated data files\2020\2020-cw.csv')  # Replace 'survey_data.csv' with your actual file path
data.columns
```

```python
graph = nx.Graph()
for _, row in data.iterrows():
    student_id = row['PersonNumber']
    semester = row['Semester']
    activity = row['Name of School:']
    graph.add_edge(student_id,semester)
pos = nx.spring_layout(graph)  # Layout algorithm for node positioning
nx.draw_networkx(graph, pos, with_labels=True,font_size=4, node_color='skyblue', node_size=500, edge_color='gray')
plt.title('Professional Development Network Graph')
plt.show()
```
